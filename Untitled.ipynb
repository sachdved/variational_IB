```python
beta = 0.1
encoder = Encoder(
    input_dim=28**2,
    hidden_dims=[1024,1024,1024],
    latent_dim=256,
    activation=torch.nn.ReLU()
)

decoder = Decoder(
    latent_dim=256,
    hidden_dims=[1024,1024,1024],
    output_dim=10,
    activation=torch.nn.ReLU()
)

vib = VariationalIB(encoder, decoder, latent_dim=256, input_dim=28**2)
optim = torch.optim.SGD(params=vib.parameters(), lr=1e-2)

for epoch in tqdm(range(10000)):
    optim.zero_grad()
    
    input_batch = {'X':batch[0].squeeze().reshape(32,784), 'Y':batch[1]}
    
    input_batch = vib(input_batch)

    recon_loss = torch.nn.CrossEntropyLoss(reduction='mean')(input_batch['Yhat'], input_batch['Y'])
    elbo_loss = vib.prior_loss(input_batch).mean()
    loss = recon_loss + 0.000001*elbo_loss
    loss.backward()
    optim.step()

    print(loss.item())
```

```python
beta = 0.1
encoder = Encoder(
    input_dim=28**2,
    hidden_dims=[1024,1024,1024],
    latent_dim=256,
    activation=torch.nn.ReLU()
)

decoder = Decoder(
    latent_dim=256,
    hidden_dims=[1024,1024,1024],
    output_dim=10,
    activation=torch.nn.ReLU()
)

vib = VariationalIB(encoder, decoder, latent_dim=256, input_dim=28**2)
optim = torch.optim.SGD(params=vib.parameters(), lr=1e-2)

I_XZ = []
I_YZ = []
for beta in betas:
    encoder = Encoder(
        input_dim=28**2,
        hidden_dims=[1024,1024],
        latent_dim=256,
        activation=torch.nn.ReLU()
    )
    
    decoder = Decoder(
        latent_dim=256,
        hidden_dims=[1024,1024],
        output_dim=10,
        activation=torch.nn.ReLU()
    )
    
    vib = VariationalIB(encoder, decoder, latent_dim=256, input_dim=28**2)
    optim = torch.optim.Adam(params=vib.parameters(), lr=1e-5)


    for epoch in tqdm(range(200)):
        elbo_loss_tot = 0
        recon_loss_tot = 0
        for batch in mnist_train_dataloader:
            optim.zero_grad()
            
            input_batch = {'X':batch[0].squeeze().reshape(batch[0].shape[0],784), 'Y':batch[1]}
            
            input_batch = vib(input_batch)
        
            recon_loss = torch.nn.CrossEntropyLoss(reduction='mean')(input_batch['Yhat'], input_batch['Y'])
            elbo_loss = vib.prior_loss(input_batch).mean()
            loss = recon_loss + 10e-6 * elbo_loss
            loss.backward()
            optim.step()
        elbo_loss_tot+=elbo_loss.item()
        recon_loss_tot+=recon_loss.item()

    i_xz = 0
    i_yz = 0

    with torch.no_grad():
        for batch in mnist_test_dataloader:
            input_batch = {'X':batch[0].squeeze().reshape(batch[0].shape[0],784), 'Y':batch[1]}
            input_batch = vib(input_batch)
            i_yz += torch.nn.CrossEntropyLoss(reduction='sum')(input_batch['Yhat'], input_batch['Y']).item()
            i_xz += vib.prior_loss(input_batch).sum().item()
    I_XZ.append(i_xz/len(mnist_testset))
    I_YZ.append(np.log(10)-i_yz/len(mnist_testset))
    print(I_XZ)
    print(I_YZ)
```
