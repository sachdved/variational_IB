```python
import pandas as pd
import numpy as np
import scipy as sp
import matplotlib.pyplot as plt
```

```python
high_diversity_relative_abundance = pd.read_csv('data/highDiversity/highDiversityRelativeAbds.csv')
```

```python
mid_diversity_relative_abundance = pd.read_csv('data/midDiversity/midDiversityRelativeAbds.csv')
```

```python
low_diversity_relative_abundance = pd.read_csv('data/lowDiversity/lowDiversityAbsoluteAbds.csv')
```

```python
plt.plot(
    np.sort(
        np.linalg.eig(high_diversity_relative_abundance.corr()).eigenvalues
    )[::-1]
    ,
    label = "Eigenvalues, High Relative Abundances"
)

plt.plot(
    np.sort(
        np.linalg.eig(mid_diversity_relative_abundance.corr()).eigenvalues
    )[::-1]
    ,
    label = "Eigenvalues, Mid Relative Abundances"
)

plt.plot(
    np.sort(
        np.linalg.eig(low_diversity_relative_abundance.corr()).eigenvalues
    )[::-1]
    ,
    label = "Eigenvalues, Low Relative Abundances"
)
plt.legend()
plt.xlabel('Eigenvalue index')
plt.ylabel('Eigenvalues')
```

```python
def entropy_calc(df):
    p = df.sum()/df.shape[0]
    return - ( p * np.log(p+1e-10) + (1-p) * np.log(1-p+1e-10)).sum()

print(entropy_calc(high_diversity_presence))
print(entropy_calc(mid_diversity_presence))
print(entropy_calc(low_diversity_presence))
```
